# Quick look at Theano

```python
import theano 
import theano.tensor as tt
```

```python
a = tt.scalar('a') # default use float64
x = tt.vector('x')
y = tt.ivector('y') # integer vector
```

```python
with pm.Model() as model:
    z = pm.Normal('z', mu=0., sigma=5.)             # ==> pymc3.model.FreeRV, or theano.tensor with logp
    x = pm.Normal('x', mu=z, sigma=1., observed=5.) # ==> pymc3.model.ObservedRV, also has logp properties
```

```python
x.logp({'z': 2.5})                                  # ==> -4.0439386
```

```python
model.logp({'z': 2.5})  
```

# use variable

```python
my_model = pm.Model()
with my_model:
    z1 = pm.Normal('z', mu=0., sigma=5.)             # ==> pymc3.model.FreeRV, or theano.tensor with logp
    x1 = pm.Normal('x', mu=z1, sigma=1., observed=5.) # ==> pymc3.model.ObservedRV, also has logp properties
x1.logp({'z': 2.5})  
```
